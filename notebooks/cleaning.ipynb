Now, let's convert the categorical variables into numerical values

Let's handle scholarship dataset

```python
df['scholarship'] = df['scholarship'].map({'Yes':1, 'No':0})
```

```python
df.head()
```

let's handle 'cgpa' feature

```python
df['cgpa'].unique()
```

cgpa values are in categorical ranges, which is fine — but to use them in a model, we'll need to convert them into numeric values.  
Ordinal Encoding (best if there's a clear order — which there is)

```python
cgpa_map = {
    'Below 2.50': 0,
    '2.50 - 2.99': 1,
    '3.00 - 3.39': 2,
    '3.40 - 3.79': 3,
    '3.80 - 4.00': 4,
    'Other': -1
}

df['cgpa_encoded'] = df['cgpa'].map(cgpa_map)
```

now, we will drop old cgpa feature

```python
df.drop(columns={'cgpa'}, axis=1, inplace=True)
```

```python
df.head()
```

```python
df['academic_year'].unique()
```

Those are values from your academic_year column. To use them in a model, it's best to convert them to numeric (ordinal) values since they represent a clear progression.

```python
year_map = {
    'First Year or Equivalent': 1,
    'Second Year or Equivalent': 2,
    'Third Year or Equivalent': 3,
    'Fourth Year or Equivalent': 4,
    'Other': -1
}

df['academic_year_encoded'] = df['academic_year'].map(year_map)
```

```python
df.head()
```

Now, drop the old one

```python
df.drop(columns={'academic_year'}, axis=1, inplace=True)
```

Now, handle the gender

```python
df['gender'].unique()
```

```python
gender_map = {
    'Male': 0,
    'Female': 1,
    'Prefer not to say': -1
}

df['gender_encoded'] = df['gender'].map(gender_map)
```

```python
df.head()
```

```python
df.drop(columns={'gender'}, axis=1, inplace=True)
```

```python
df.head()
```

Handle the feature 'age' now

```python
df['age'].unique()
```

That's your age column — stored as categorical age ranges.  
These ranges clearly follow an order, so map them ordinally

```python
age_map = {
    'Below 18': 0,
    '18-22': 1,
    '23-26': 2,
    '27-30': 3,
    'Above 30': 4
}

df['age_encoded'] = df['age'].map(age_map)
```

```python
df.head()
```

```python
df.age_encoded.unique()
```

```python
df.drop(columns={'age'}, axis=1, inplace=True)
```

```python
df.head()
```

Now, let's handle our dependent variable, i.e., depression_label

```python
df['depression_label'].unique()
```

```python
label_map = {
    'No Depression': 0,
    'Minimal Depression': 1,
    'Mild Depression': 2,
    'Moderate Depression': 3,
    'Moderately Severe Depression': 4,
    'Severe Depression': 5
}

df['depression_label_encoded'] = df['depression_label'].map(label_map)
```

```python
df.head()
```

```python
df.drop(columns={'depression_label'}, axis=1, inplace=True)
```
